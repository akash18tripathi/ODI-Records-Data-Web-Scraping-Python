```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np
from collections import OrderedDict
```

```python
players =[]
year = []
runs=[]
nationality = []
```

```python
for _ in range(1,215):
    soup = BeautifulSoup(requests.get("http://stats.espncricinfo.com/ci/engine/stats/index.html?class=2;home_or_away=1;home_or_away=2;home_or_away=3;page="+str(_)+";result=1;result=2;result=3;result=5;template=results;type=batting;view=year").text)
    
    for i in range(0,50):
        players.append(soup.find_all('tr',attrs={'class':'data1'})[i].find_all('td')[0].a.string)
        year.append(soup.find_all('tr',attrs={'class':'data1'})[i].find_all('td')[12].string)
        runs.append(soup.find_all('tr',attrs={'class':'data1'})[i].find_all('td')[4].string)
        nationality.append(soup.find_all('tr',attrs={'class':'data1'})[i].find_all('td')[0].a.next_sibling)
```

```python
for i in list(zip(players,nationality,runs,year)):
    print(i[0],i[1],i[3],i[2])
```

```python
df = pd.DataFrame(list(OrderedDict.fromkeys(players)),columns=['players'])
```

```python
df.index = df.players
```

```python
df = df.drop('players',axis=1)
```

```python
df.head()
```

```python
for i in list(zip(players,nationality)):
    df.loc[i[0],'Nationality']=i[1]
```

```python
for i in list(zip(players,runs,year)):
    df.loc[i[0],i[2]]=i[1]
```

```python
df.head()
```

```python
df = df.set_index(['players','Nationality'])
```

```python
df.head()
```

```python
df = df.sort_index(axis=1)
```

```python
df = df.replace(np.nan,0)
```

```python
df = df.apply(pd.to_numeric)
```

```python
df = df.cumsum(axis=1)
```
